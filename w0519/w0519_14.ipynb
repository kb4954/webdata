```python
import pandas as pd

df = pd.read_csv('data/stuscore.csv',index_col='번호')
df
```

str 함수

```python
### a가 들어가 있는 문자를 검색하시오.

df[df['이름'].str.contains('a')].count() # 40개
df[df['이름'].str.contains('A')] # 8개
df[df['이름'].str.contains('A')|df['이름'].str.contains('a')].count() # 48개
```

```python
df = pd.read_excel('data/score.xlsx',index_col='지원번호')

# 홍이 포함되어 있는 학생의 국어,영어,수학,과학, 성적을 출력하시오

score = df[df['이름'].str.contains('홍')]
score[['이름','국어','영어','수학','과학']]
# 슬라이싱도 가능 -> 슬라이싱할때는 슬라이싱만 해야됨. 아니면 다 쉼표
## row를 기준으로 출력
# df.loc[df['이름'].str.contains('홍')]
```

```python
## SW특기가 python, java 글자가 포함되어 있는 학생의 이름,학교,키,SW특기를 출력하시오.
sw = df[df['SW특기'].str.lower().str.contains('python')|df['SW특기'].str.contains('Java')]
sw[['이름','학교','키','SW특기']]
```

소문자로 변경 후 글자가 있는지 검색

```python
df['이름'].str.lower() # 모든 글자 소문자로 치환

# 모든 글자를 소문자로 변경 후 a가 포함되어 있는 문자 검색(?)
df[df['이름'].str.lower().str.contains('a')]

# 모든 글자 대문자로 변경
df[df['이름'].str.upper().str.contains('A')] #(?)
df['이름'].str.upper()
```

```python
### endswith - 해당글자로 끝나는 문자 검색
df[df['이름'].str.endswith('신')]
df[df['이름'].str.endswith('n')]
```

```python
# contains - 해당 글자가 포함되어 있는 문자 검색

df[df['이름'].str.contains('s')]
df[df['이름'].str.contains('순')]
df[~df['이름'].str.contains('순')] #~not을 의미
```
