데이터 추가

```python
## 컬럼 추가
# 없는 컬럼을 입력하면 컬럼이 추가 됨.
df['합계'] = df['국어']+df['영어']+df['수학']+df['과학']+df['사회']
df

# 평균을 추가하시오.
df['평균'] = (df['국어']+df['영어']+df['수학']+df['과학']+df['사회'])/5
df

## 결과 컬럼을 추가하고 내용은 불합격입력하시오.
df['결과'] = "불합격"
df

## 평균컬럼이 60점 이상인 학생만 출력하시오.
avg = df['평균']>=60
df[avg]

# 평균이 60점 이상이면 결과 컬럼에 합격이라고 입력하시오.
df[avg]['결과']
df.loc[avg,'결과']='합격'
df
```
