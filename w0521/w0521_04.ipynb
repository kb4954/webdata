```python
## 90점 이상 A, 80점 이상 B, 70점 이상 C, 60점이상 D, 60점 미만 F
def rank(score):
    result = ""
    if 90>=score: result = "A"
    if 80>=score: result = "B"
    if 70>=score: result = "C"
    if 60>=score: result = "D"
    else: result = "F"
    return result
    
    
df['등급'] = df['평균'].apply(rank)
df
```

```python
### 함수를 사용해서 요일 0:일요일, 1:월, 2:화, 3:수, 4:목, 5:금, 6:토 변경하시오
def week(wData):
    for i in range(8):
        result = ""
        if wData==0: result = "일요일"
        if wData==1: result = "월요일"
        if wData==2: result = "화요일"
        if wData==3: result = "수요일"
        if wData==4: result = "목요일"
        if wData==5: result = "금요일"
        else: result = "토요일"
        return result

df['요일'] = df['요일'].apply(week)
df
```
