```python
# 리스트 타입만 들어오면 어떤 정보든 가능
# label: 범례
# 선형태 - linewidth : 선의 두께
# linestyle='None' ls(축약버전): 선제거 
# : - 점선
# --: 대쉬선 
# -.:점선 점 형태

# 마커 형태
# marker: 마커표시 - o,v,x,s 
# markeredgecolor mec:마커 테두리 
# markerfacecolor mfc: 마커 테두리 색상

## 'ro--' : color red, 마커모양'o',선의 스타일 --
plt.plot(x,y,'ro--',label = '서울', linewidth=5)
# plt.plot(x,y,label = '서울',linewidth=2,ls='-.',color = 'red', alpha = 0.7, 
#          mec='red', mfc='yellow',marker="o",markersize=10) 
# plt.plot(x,y,label = '서울',linewidth=2,marker="o",markersize=10, 
#          markeredgecolor='red', markerfacecolor='yellow') 
# plt.plot(df['이름'],df['국어']) 


## legend # 주로 코드 밑에 있어야지 나타남.
# loc 범례위치 - y축: upper,center,lower  x축: left,center,right

plt.title('그래프') # 상단 제목 입력
plt.legend(loc=(0.25,0.25)) # 위치점 지정(0.5,0.5) : 가운데 지정
plt.show()
```

```python
plt.plot(df['이름'],df['국어'],label = '국어',marker="o")
plt.plot(df['이름'],df['영어'],label = '영어',marker="v",alpha = 0.2)
plt.plot(df['이름'],df['수학'],label = '수학',marker="x",alpha = 0.2)
plt.title("학생성적그래프")
plt.legend()
plt.show()
```
