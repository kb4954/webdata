```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # windows 한글 폰트 연결
# matplotlib.rcParams['font.family'] = 'AppleGothic' # mac
matplotlib.rcParams['font.size'] = 15                # 폰트사이즈
matplotlib.rcParams['axes.unicode_minus'] = False    # 그래프 마이너스  표시
```

```python
x = [1,2,3]
y = [2,4,8]
```

해상도 지정<br>
plt.figure(dpi=72,300)

```python
# x축,y축으로 크기 조정
# plt.figure : plt.plot 위쪽에 있어야 적용됨.
# dpi: 파일 저장할때 해상도를 결정 - 파일크기가 달라짐.
# facecolor : 배경색을 지정
# figsize: 화면 사이즈 변경
plt.figure(figsize=(10,5),facecolor='yellow')
plt.plot(x,y,label='온도',marker='o',markersize=7)
plt.title('그래프')
plt.legend()
# plt.savefig('a1.png') # plt.plot 아래쪽에 위치해야함.
plt.show()
```

```python
plt.plot(x,y,label='온도',marker='o',markersize=7)
plt.title('그래프')
plt.legend()
## 선 위에 글자를 입력 plt.text
## text(x축,y축,데이터값,글자위치,글자색상)
plt.text(x[0],y[0]+0.2,2,ha='center',color='red')
plt.text(x[1],y[1]+0.2,4,ha='center',color='red')
plt.text(x[2],y[2]+0.2,8,ha='center',color='red')
plt.show()
```
