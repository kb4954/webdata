```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # windows 한글 폰트 연결
# matplotlib.rcParams['font.family'] = 'AppleGothic' # mac
matplotlib.rcParams['font.size'] = 15                # 폰트사이즈
matplotlib.rcParams['axes.unicode_minus'] = False    # 그래프 마이너스  표시
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y,marker='o',markersize=7,
         mec='red', mfc='yellow', label='점수',
         ls='--')
plt.title('성적그래프')
plt.legend()
plt.show()
```

```python
plt.plot(x,y,'bo--',label='점수',mfc='r')
plt.title('성적그래프')
plt.xlabel('월별')
plt.ylabel('점수')
plt.xticks([1,2,3])
plt.yticks([8,4,8])
plt.legend()
plt.show()
```

plt.text() : 그래프에 글자를 입력할때

```python
x =[1,2,3]
y =[2,4,8]
plt.plot(x,y,marker='o')
# for i,val in enumerate(y):
#     plt.text(x[i]+0.02,y[i]+0.2,val,ha='center')
plt.text(x[0],y[0]+0.2,2,ha='center')
plt.text(x[1],y[1]+0.2,4,ha='center')
plt.text(x[2],y[2]+0.2,8,ha='center')
plt.show()
```

```python
x = df['지원번호']
y= df['국어']
plt.plot(df['지원번호'],df['국어'],marker='o')
plt.plot(df['지원번호'],df['영어'],marker='o')
plt.yticks([10,30,50,70,90,100]) # text가 선 밖으로 벗어났을때 간격을 줘서 안으로 넣을 수 있음.
# plt.text(x[0],y[0],90,ha='center')
for i,v in enumerate(y):
    plt.text(df['지원번호'][i],df['국어'][i]+2,v,ha='center')
plt.show()
```

```python
y = df['영어']
plt.plot(df['지원번호'],df['영어'],marker='o')
plt.yticks([10,30,50,70,90,100])
for i,v in enumerate(y):
    plt.text(df['지원번호'][i],df['영어'][i]-6,v,ha='center')
plt.show()
```
