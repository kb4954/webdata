```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # windows 한글 폰트 연결
# matplotlib.rcParams['font.family'] = 'AppleGothic' # mac
matplotlib.rcParams['font.size'] = 15                # 폰트사이즈
matplotlib.rcParams['axes.unicode_minus'] = False    # 그래프 마이너스  표시
```

```python
day = [1,2,3]
kor = [100,80,90]
eng = [95,85,92]
math = [80,90,70]
```

여러개 그래프 그리기

```python
plt.plot(day,kor,label='국어',marker='o',color='#b96aeb')
plt.plot(day,eng,label='영어',marker='v',ls='--',color='#87b1ff')
plt.plot(day,math,label='수학',marker='x',ls=':',color='#1b9119')
plt.yticks([60,70,80,90,100,110])
plt.xticks([1,2,3,3.2])
plt.legend(loc='upper center',ncol=3) # ncol = 열을 3개로 나눠서 총 1개의 행으로
for i,v in enumerate(kor):
    plt.text(day[i],kor[i]+1,v,ha='center')
    plt.text(day[i],eng[i]+2,eng[i],ha='center')
    plt.text(day[i],math[i]+1.8,math[i],ha='center')
plt.show()
```

```python
name = ['홍길동','유관순','이순신']
# no = [1,2,3]
no = np.arange(1,4) # numpy 배열 형태로 적용
kor = [100,80,90]
eng = [95,85,92]
math = [80,90,77]
```

막대그래프 이동

```python
colors = ['r','g','b']
## width: 바의 사이즈 조절
## no numpy 배열 형태
w = 0.2
# bar(x좌표, y좌표)
bar = plt.bar(no-w,kor,width=0.2)
# set_hatch : bar에 무늬를 넣을 수 있음.
bar[0].set_hatch('/') 
bar[1].set_hatch('x') 
bar[2].set_hatch('..') 

plt.bar(no,eng,width=0.2)
plt.bar(no+w,math,width=0.2)
plt.ylim(70,105)
plt.xticks(no,name) # 쓰는건 no지만 표시는 name으로 하라는 뜻
## 각각의 막대그래프 위에 점수를 출력하시오.
for i,v in enumerate(kor):
    plt.text(no[i]-w,kor[i]-3,v,ha='center')
    plt.text(no[i],eng[i]-3,eng[i],ha='center')
    plt.text(no[i]+w,math[i]-2,math[i],ha='center')

plt.show()

```

```python
name = ['홍길동','유관순','이순신']
# no = [1,2,3]
no = np.arange(1,4) # numpy 배열 형태로 적용
kor = [100,80,90]
eng = [95,85,92]
math = [80,90,77] 
```

```python
## 3개의 막대 그래프를 출력하시오. grid: -- 색상은 검정 투명도 0.1
w = 0.2
plt.grid(color='black',alpha=0.1, ls='--')
plt.bar(no-w,kor,width=0.2)
plt.bar(no,eng,width=0.2)
plt.bar(no+w,math,width=0.2)
plt.ylim(70,105)
plt.xticks(no,name)
for i,v in enumerate(kor):
    plt.text(no[i]-w,kor[i]-2,v,ha='center')
    plt.text(no[i],eng[i]-2,eng[i],ha='center')
    plt.text(no[i]+w,math[i]-2,math[i],ha='center')
plt.show()
```
